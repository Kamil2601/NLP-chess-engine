```python
import requests
import re


headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

root_url = "https://www.chessgames.com"

# game_regex = 'href="/nodejs/game/downloadGamePGN/lasker_chigorin_1895.pgn?gid=1036398"'
game_regex = 'href="/nodejs/game/downloadGamePGN/.+.pgn\?gid=\d+"'

def get_pgn_url(url):
    try:
        # Send a GET request to the URL
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print("Error: " + url)
            return None

        res = re.search(game_regex, response.text)
        return res
    except Exception as e:
        print("Error: " + url)
        return None



# Example usage:
url = "/perl/chessgame?gid=1489285"  # Replace this with your desired URL
pgn_url = get_pgn_url(url)
# response = requests.get(pgn_url, headers=headers)
pgn_url
```

```python

regex = "/perl/chessgame\?gid=\d+"

games_urls = []

for page in range(1, 127):
    url = f"https://www.chessgames.com/perl/chess.pl?page={page}&playercomp=either&annotated=1"
    response = requests.get(url, headers=headers)

    print(response)
    
    urls = re.findall(regex, response.text)

    games_urls.extend(urls)

games_urls = list(set(games_urls))
```

```python
with open('urls.txt', 'w') as f:
    for item in games_urls:
        f.write(f"{root_url}{item}\n")
```
