```python
table_names = ["angelfire_moves", "chessbase_moves", "chessgames_moves", "gameknot_moves", "github_ValdemarOrn_Chess_moves", "lichess_studies_moves", "scrapped_moves"]
```

```python
moves_dfs = [load_sql_to_df(f"SELECT * FROM {table_name}", "../../data/chess_2.db") for table_name in table_names]
```

```python
moves_dfs = [df.drop(columns=["index"]) for df in moves_dfs]
```

```python
moves_df = pd.concat(moves_dfs)
```

```python
moves_df.comment = moves_df.comment.apply(lambda com: com.strip())
moves_df
```

```python
moves_df = moves_df[(moves_df.nags != "[]") | (moves_df.comment != "")]
```

```python
save_to_sql(moves_df, "../../data/chess_moves_comments_nags.db", "all_moves", if_exists='replace')
```

```python
moves_with_comments = moves_df[moves_df.comment != ""]
```

```python
moves_with_comments
```

```python
moves_with_comments_labeled = moves_with_comments[moves_with_comments.sentiment != -1]
moves_with_comments_labeled
```

```python
save_to_sql(moves_with_comments_labeled, "../../data/chess_moves_comments_nags.db", "labeled_moves", if_exists='replace')
```

```python
moves_with_comments_unlabeled = moves_with_comments[moves_with_comments.sentiment == -1]
save_to_sql(moves_with_comments_unlabeled, "../../data/chess_moves_comments_nags.db", "unlabeled_moves", if_exists='replace')
```

```python
nags_without_comments = moves_df[moves_df.comment == ""]
save_to_sql(nags_without_comments, "../../data/chess_moves_comments_nags.db", "nags_without_comments", if_exists='replace')
```
